```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(10, 20)
y = np.array([2, 1, 4, 5, 8, 12, 18, 25, 96, 48])

def mean(x):
    return x.sum() / x.size

def var(x):
    intmean = mean(x)
    result = 0.0
    for i in x:
        result += (i-intmean)**2
    return result / x.size

def covar(x, y):
    if(x.size != y.size):
        return 0

    xmean = mean(x)
    ymean = mean(y)
    result = 0.0
    for i in range(x.size):
        result += (x[i] - xmean)*(y[i] - ymean)
    return result / (x.size)

def cor(x, y):
    return covar(x, y) / np.sqrt(var(x)*var(y))



print(mean(x))
print(np.mean(x))
print(var(y))
print(np.var(y))
print(covar(x, y))
print(np.cov(x, y, bias=True)[1][0])
print(cor(x, y))
print(np.corrcoef(x, y))


```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

num_samples = 400

# The desired mean values of the sample.
mu = np.array([5.0, 0.0, 10.0])

# The desired covariance matrix.
r = np.array([
        [  3.40, -2.75, -2.00],
        [ -2.75,  5.50,  1.50],
        [ -2.00,  1.50,  1.25]
    ])

# Generate the random samples.
rng = np.random.default_rng()
y = rng.multivariate_normal(mu, r, size=num_samples)


# Plot various projections of the samples.
plt.plot(y[:,1], y[:,2], 'b.', alpha=0.25)
plt.plot(mu[1], mu[2], 'ro', ms=3.5)
plt.xlabel('y[1]')
plt.axis('equal')
plt.grid(True)
```
